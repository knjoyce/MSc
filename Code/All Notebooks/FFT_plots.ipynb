# FFT - 1 MHz

```python
exec(open('ComputeDelays.py').read())
import numpy as np
```

```python
def fft(t,sig):
        
    n=np.size(t)
    dt=(t[2]-t[1])
    f=np.fft.fftshift(np.fft.fftfreq(n,dt))
    FT=np.fft.fftshift(np.fft.fft(sig))
    
    
    return f,FT
```

```python
# Aluminium
n_Aluminium=np.size(Aluminium.time)
dt_Aluminium=(Aluminium.time[2]-Aluminium.time[1])*1e6
f_Aluminium=np.fft.fftshift(np.fft.fftfreq(n_Aluminium,d=dt_Aluminium))*1000
FT_Aluminium=np.fft.fftshift(np.fft.fft(Aluminium.data - np.mean(Aluminium.data)))

n_Sandstone=np.size(Sandstone.time)
dt_Sandstone=(Sandstone.time[2]-Sandstone.time[1])*1e6
f_Sandstone=np.fft.fftshift(np.fft.fftfreq(n_Sandstone,d=dt_Sandstone))*1000
FT_Sandstone=np.fft.fftshift(np.fft.fft(Sandstone.data))


plt.plot(f_Aluminium,np.abs(FT_Aluminium), label="Aluminium")
plt.plot(f_Sandstone,np.abs(FT_Sandstone), label="Sandstone")
plt.xlim([-2000,2000])
plt.legend()
plt.xlabel('Frequency (kHz)')
plt.ylabel('Intensity')
plt.title('FFT Material Comparison - 1 MHz')
```
