Include this result in thesis: 

$width = 2f/n$

Explain that peak width increases with frequency, when other parameters are constant. The peak width is determined by the time its takes for the wave to complete n cycles at a given frequency. Lower frequencies take longer and therefore have narrower peaks. When total time is fixed, as above, the peak width can be thought as related to the ratio of how much of the total time is used for the 'actual wave' and how much is padding. The padding does not narrow the peaks, rather it causes interpolation using a sinc function.
