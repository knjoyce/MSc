```python
import numpy as np
import matplotlib.pyplot as plt

# original signal
time = np.arange(0,4*np.pi,0.1)
y = np.sin(time)

# 1. shift signal to min = 0
ymin = np.min(y)
shift = np.abs(ymin)
y_shifted = y+shift

# 2. scale signal
scale = 16
y_scaled = (scale/np.max(y_shifted))*y_shifted

# 3. round signal
y_rounded = np.round(y_scaled)

# 4. unshift & unscale
y_unscale = y_rounded/(scale/np.max(y_shifted))
y_unshift = y_unscale - shift


plt.plot(time,y, label='original')
plt.plot(time,y_shifted, label='1. shifted')
plt.plot(time,y_scaled, label='2. scaled')
plt.plot(time,y_rounded, label='3. rounded')
plt.plot(time,y_unshift, label='4. undo')

plt.legend()
plt.title('DSP Steps')

```

```python
def mysin(f,fs,tmax=0.1,tmin=0.0):
    
    #time = np.arange(0,4*np.pi,0.1)
    dt = 1/fs
    t = np.arange(tmin,tmax,dt)
    y = np.sin(2*np.pi*f*t)
    #mysin(f,fs,tmax=0.1):
    # f is the frequency of the signal
    # tmax is the final time
    # fs is the sampling frequency, fs=1/dt
    ##HINT: You need to make a time vector with the appropriate dt, so first compute that then dt, then t, then y
    return y,t
```

```python
def bases_proj(a,b):
    #bases_proj(a,b) projects vector a onto vector b (or vice-versa provided all signals are real)
    proj = np.dot(a,b) #dot product of a and b
    return proj
```
