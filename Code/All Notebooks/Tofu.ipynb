# Frequency Shift - Cont.

```python
max1=np.argmax(np.abs(FT_tofuC_100))
max3=np.argmax(np.abs(FT_tofuC_300))
max5=np.argmax(np.abs(FT_tofuC_500))
max8=np.argmax(np.abs(FT_tofuC_800))
max10=np.argmax(np.abs(FT_tofuC_1000))

shift100 = np.abs(100 - np.abs(f_tofuC_100[max1]))
shift300 = np.abs(300 - np.abs(f_tofuC_300[max3]))
shift500 = np.abs(500 - np.abs(f_tofuC_500[max5]))
shift800 = np.abs(800 - np.abs(f_tofuC_800[max8]))
shift1000 = np.abs(1000 - np.abs(f_tofuC_1000[max10]))

freqs = [100,300,500,800,1000] 
shifts = [shift100,shift300,shift500,shift800,shift1000]

plt.plot(freqs,shifts,'-o')
slope, intercept = np.polyfit(freqs,shifts,1)

#plt.xlim([-100,1500])
#plt.legend()
plt.xlabel('Frequency (kHz)')
plt.ylabel('Absolute Shift (kHz)')
plt.title('Frequency Shift - Tofu (Continuous)')
plt.savefig("TofuFreqShift_cont.jpg",dpi=300)
```

# FFT Amplitude vs Frequency

```python
max_C = np.max(np.abs(FT_tofuC_800)) # found visually from FFT_plots notebook
max_4 = np.max(np.abs(FT_tofu4_100)) # found visually from FFT_plots notebook


max_100C = np.max(np.abs(FT_tofuC_100)/max_C)
max_300C = np.max(np.abs(FT_tofuC_300)/max_C)
max_500C = np.max(np.abs(FT_tofuC_500)/max_C)
max_800C = np.max(np.abs(FT_tofuC_800)/max_C)
max_1000C = np.max(np.abs(FT_tofuC_1000)/max_C)

max_1004 = np.max(np.abs(FT_tofu4_100)/max_4)
max_3004 = np.max(np.abs(FT_tofu4_300)/max_4)
max_5004 = np.max(np.abs(FT_tofu4_500)/max_4)
max_8004 = np.max(np.abs(FT_tofu4_800)/max_4)
max_10004 = np.max(np.abs(FT_tofu4_1000)/max_4)

amplitudes_C = [max_100C,max_300C,max_500C,max_800C,max_1000C]
amplitudes_4 = [max_1004,max_3004,max_5004,max_8004,max_10004]
freqs = [100,300,500,800,1000] 

plt.plot(freqs,amplitudes_C, label="Continous")
plt.plot(freqs,amplitudes_4, label="4 Cycles")
plt.legend()
plt.xlabel('Frequency (kHz)')
plt.ylabel('Amplitude')
plt.title('FFT Amplitude vs Frequency - Tofu (Normalized to Max)')
plt.savefig("Amp_vsFreq_Tofu.jpg",dpi=300)
```
