## Extend 4 cylce data into continuous by multiplication

```python
exec(open('ComputeDelays.py').read())
import matplotlib.pyplot as plt

GS_4=timedata('GS_4')
readlabdat('jun27_2.csv',GS_4)

plt.plot(GS_4.time, GS_4.data, '-o')
```

## Crop data to remove zeros

```python
short_time =GS_4.time[200:1000]
short_data = GS_4.data[200:1000]
plt.plot(short_time, short_data)
```

```python
n_1=np.size(short_time)
dt_1=(short_time[2]-short_time[1])
f_1=np.fft.fftshift(np.fft.fftfreq(n_1,d=dt_1))
#FT_=np.fft.fftshift(np.fft.fft(y))
FT_1=np.fft.fftshift(np.fft.fft(short_data - np.mean(short_data)))

plt.plot(f_1/1000,np.abs(FT_1),'-o', label="500 kHz")

plt.xlim([0,1400])
plt.legend()
plt.xlabel('Frequency (kHz)')
plt.ylabel('Intensity')
plt.title('FFT Generator to Oscilliscope')
```

```python
time = short_time
data = np.array(short_data)

data_long = np.insert(data,len(data), data)

zeros = len(time)
newtimes = np.zeros(zeros)
for i in range(1,len(time) + 1):   
    newtimes[i-1]= time[len(time)-1] + i*dt_1

time_long = np.append(time,newtimes)

plt.plot(time_long, data_long)
```

```python
n_2=np.size(time_long)
dt_2=(time_long[2]-time_long[1])
f_2=np.fft.fftshift(np.fft.fftfreq(n_2,d=dt_2))
#FT_=np.fft.fftshift(np.fft.fft(y))
FT_2=np.fft.fftshift(np.fft.fft(data_long - np.mean(data_long)))

plt.plot(f_2/1000,np.abs(FT_2),'-o', label="doubled")
plt.plot(f_1/1000,np.abs(FT_1),'-o', label="original")

plt.xlim([0,1400])
plt.legend()
plt.xlabel('Frequency (kHz)')
plt.ylabel('Intensity')
plt.title('FFT Generator to Oscilliscope')
```

```python
data_long = np.array(data_long)
data_long2 = np.insert(data_long,len(data_long), data_long)

zeros = len(time_long)
newtimes = np.zeros(zeros)
for i in range(1,len(time_long) + 1):   
    newtimes[i-1]= time_long[len(time_long)-1] + i*dt_1

time_long2 = np.append(time_long,newtimes)

plt.plot(time_long2, data_long2)
```

```python
n_3=np.size(time_long2)
dt_3=(time_long2[2]-time_long2[1])
f_3=np.fft.fftshift(np.fft.fftfreq(n_3,d=dt_3))
#FT_=np.fft.fftshift(np.fft.fft(y))
FT_3=np.fft.fftshift(np.fft.fft(data_long2 - np.mean(data_long2)))

plt.plot(f_1/1000,np.abs(FT_1),'-o', label="original")
plt.plot(f_2/1000,np.abs(FT_2),'-o', label="x2")
plt.plot(f_3/1000,np.abs(FT_3),'-o', label="x4")

plt.xlim([0,1400])
plt.legend()
plt.xlabel('Frequency (kHz)')
plt.ylabel('Intensity')
plt.title('FFT Generator to Oscilliscope - extended data')
```

# repeat above to plot for group meeting

```python
Tofu_4=timedata('Tofu_4')
readlabdat('jul13_8.csv',Tofu_4)

plt.title('cropped data')
plt.plot(Tofu_4.time[581:699], Tofu_4.data[581:699])
```

```python
dt = Tofu_4.time[2] - Tofu_4.time[1]
nd = np.size(Tofu_4.data[581:699])
time2=np.arange(0,nd*4)*dt
time=np.arange(0,nd)*dt

data = np.array(Tofu_4.data[581:699])
data2 = np.insert(data,len(data), data)
data2= np.insert(data2, len(data2),data2)

plt.title('extended data')
plt.plot(time2, data2)
```

```python
n=np.size(Tofu_4.time[581:699])
f=np.fft.fftshift(np.fft.fftfreq(n,d=dt))
FT=np.fft.fftshift(np.fft.fft(data - np.mean(data)))

n2=np.size(time2)
f2=np.fft.fftshift(np.fft.fftfreq(n2,d=dt))
FT2=np.fft.fftshift(np.fft.fft(data2 - np.mean(data2)))

plt.plot(f/1000,np.abs(FT),'-o', label="Original (df=168 kHz)")
plt.plot(f2/1000,np.abs(FT2),'-o', label="x4 (df=42 kHz)")

plt.xlim([0,1400])
plt.legend()
plt.xlabel('Frequency (kHz)')
plt.ylabel('Intensity')
plt.title('Tofu FFT repeated data')
```

```python
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(7, 7))

# plot time signal:
axs[0].set_title("Time Domain")
axs[0].plot(time2*1e6, data2, label="x4")
axs[0].plot(time*1e6, Tofu_4.data[581:699], label="Original")
axs[0].set_xlabel("Time ($\mu$s)")
axs[0].set_ylabel("Amplitude (V)")
axs[0].legend()

axs[1].set_title("Frequency Domain")
axs[1].plot(f2/1000,np.abs(FT2)/max(np.abs(FT2)),'-o', label="x4 (df~42 kHz)")
axs[1].plot(f/1000,np.abs(FT)/max(np.abs(FT)),'-o', label="Original (df~168 kHz)")

axs[1].set_xlabel("Frequency (kHz)")
axs[1].set_ylabel("Intensity")
axs[1].set_xlim([0, 1000])
plt.legend()

fig.tight_layout()

plt.savefig('extending_data_tofu.jpg', dpi=400)
```
